```python
dt = dict()

arr = os.listdir("/Users/akshitasaxena/Downloads/Driver Drowsiness Dataset (DDD) (1) 2/Non Drowsy")
arr.sort()

for filename in arr:
    
    lt = filename[0]
    
    if lt not in used_list_alert :
        if lt not in dt :
            dt[lt] = 1
            folder_name = f"{name_str}{3 + len(dt)}"

        if dt[lt] <= 26:
            shutil.copy(f"/Users/akshitasaxena/Downloads/Driver Drowsiness Dataset (DDD) (1) 2/Non Drowsy/{filename}", folder_name)
            dt[lt] += 1
            
        print(filename)
    
print(dt)
```
